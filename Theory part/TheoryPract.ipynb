### 4. Find the spectrum of the following (a) a line with slope 1. (b) alternate black and white horizontal lines. (c) alternate black and white vertical lines.

```python
row = 150
col = 150
lineimage = np.zeros(row*col).reshape(row, col)
```

```python
for i in range(row):
    for j in range(col):
        if i==j:
            lineimage[i][j] = 1
```

```python
fft_line = np.fft.fft2(lineimage)
fftshift_line = np.fft.fftshift(fft_line)

plt.subplot(1, 3, 1)
plt.imshow(lineimage, cmap='gray')
plt.title('Line')

plt.subplot(1, 3, 2)
plt.imshow(abs(fft_line))
#plt.colorbar()
plt.title('FFT Line')

plt.subplot(1, 3, 3)
plt.imshow(abs(fftshift_line))
#plt.colorbar()
plt.title('FFT SHIFT Line')

plt.tight_layout()
plt.show()
```

```python
row = 20
col = 20
verticalStrips = np.zeros(row*col).reshape(row, col)

for i in range(row):
    for j in range(col):
        if j%2 == 0:
            verticalStrips[i][j] = 1
            
fft_vertical = np.fft.fft2(verticalStrips)
fftshift_vertical = np.fft.fftshift(fft_vertical)

plt.subplot(1, 3, 1)
plt.imshow(verticalStrips, cmap='gray')
plt.title('Vertical')

plt.subplot(1, 3, 2)
plt.imshow(abs(fft_vertical))
#plt.colorbar()
plt.title('FFT Vertical')

plt.subplot(1, 3, 3)
plt.imshow(abs(fftshift_vertical))
#plt.colorbar()
plt.title('FFT SHIFT Vertical')

plt.tight_layout()
#plt.colorbar()
plt.show()
```

```python
row = 25
col = 25
horizStrips = np.zeros(row*col).reshape(row, col)

for i in range(row):
    for j in range(col):
        if i%2 == 0:
            horizStrips[i][j] = 1
            
fft_horiz = np.fft.fft2(horizStrips)
fftshift_horiz = np.fft.fftshift(fft_horiz)

plt.subplot(1, 3, 1)
plt.imshow(horizStrips, cmap='gray')
plt.title('Horizontal')

plt.subplot(1, 3, 2)
plt.imshow((abs(fft_horiz)))
#plt.colorbar()
plt.title('FFT Horizontal')

plt.subplot(1, 3, 3)
plt.imshow(abs(fftshift_horiz))
#plt.colorbar()
plt.title('FFT SHIFT Horizontal')

plt.tight_layout()
plt.show()
```

### 5. Say, you have an image having a black background with a white rectangle at the upper left corner. Is it possible to express the spectrum of an image which has the same square sitting at the center in terms of spectrum of the first image? If yes, how? If no, why?

```python
row = 12
col = 12
blockimage = np.zeros(row*col).reshape(row, col)

for i in range(int(row/4)):
    for j in range(int(col/2)):
        blockimage[i][j] = 1
 
plt.subplot(1, 3, 1)
plt.imshow(blockimage, cmap='gray')
plt.title('blockimage')
#plt.show()

fft_block = np.fft.fft2(blockimage)
fftshift_block = np.fft.fftshift(fft_block)

plt.subplot(1, 3, 2)
plt.imshow(abs(fft_block))
plt.title('fft_block')
#plt.show()

plt.subplot(1, 3, 3)
plt.imshow(abs(fftshift_block))
plt.title('fftshift_block')
plt.tight_layout()
plt.show()
```

```python
row = 12
col = 12
blockimage2 = np.zeros(row*col).reshape(row, col)


for i in range(int(row/2)-int(row/8), int(row/2)+int(col/8)):
    for j in range(int(col/2)-int(col/4), int(col/2)+int(col/4)):
        blockimage2[i][j] = 1

plt.subplot(1, 3, 1)
plt.imshow(blockimage2, cmap='gray')
plt.title('blockimage2')
#plt.show()

fft_block2 = np.fft.fft2(blockimage2)
fftshift_block2 = np.fft.fftshift(fft_block2)

plt.subplot(1, 3, 2)
plt.imshow(abs(fft_block2))
plt.title('fft_block2')
#plt.show()

plt.subplot(1, 3, 3)
plt.imshow(abs(fftshift_block2))
plt.title('fftshift_block2')
plt.tight_layout()
plt.show()
```
